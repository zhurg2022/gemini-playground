# Processing after download

```python
#@title Write a bash for processing
com_file = open('1.sh','w')
com_file.write('''
#!/bin/bash

# 进入每个子文件夹
for dir in */; do
  cd "$dir"
  # 进入内部文件夹
  for inner_dir in */; do
    cd "$inner_dir"
    # 解压所有.gz文件
    for file in *.gz; do
      gunzip "$file"
    done
    # 复制所有解压后的文件到上层文件夹
    mv *.sdf ../../
    cd ../
  done
  cd ../
done

# 合并所有文件
cat *.sdf > all_zinc.sdf
''')
com_file.close()
```

```python
#@title Run bash
!sh 1.sh
```
